```
import pandas as pd
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from textblob import TextBlob as tb
```

```
data = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz", lines = True) #Reading the Office Products file 
```

```
df = data.copy()
```

```
line = df['reviewText'][0] #storing the string as 'line'
test = df['reviewText'][0:50]
test = pd.DataFrame(test)
test.head()
```

```
from string import punctuation
def strip_punctuation(s):
    return ''.join(c for c in s if c not in punctuation)
```

```
n_line = (strip_punctuation(line))
```

```
def clean_data(str):
  str = str.lower()
  n_str = strip_punctuation(str)
  stopword = set(stopwords.words('english')) 
  word_tokens = nltk.word_tokenize(n_str)
  removing_stopwords = [word for word in word_tokens if word not in stopword]
  return removing_stopwords
```

```
clist = clean_data(line)
```

```
def getUniqueWords(list) :
    uniqueWords = [] 
    for i in list:
        if not i in uniqueWords:
            uniqueWords.append(i)
    return uniqueWords
```

```
flist = getUniqueWords(clist)
ctest = test.reviewText.apply(clean_data)
ctest = pd.DataFrame(ctest)
```

```
ftest = ctest.reviewText.apply(getUniqueWords)
sentences = list(ftest)
words = [[i] for i in flist]
words = list(words)
```

```
from itertools import combinations
```

```
def combination(lists,pairs):
     comb = combinations(lists,pairs)
     for word in list(comb):
        print(word)
```

```
import gensim
from gensim import interfaces, utils, matutils
from gensim.matutils import dirichlet_expectation, mean_absolute_difference
from gensim.models import Word2Vec
from gensim.test.utils import common_texts, get_tmpfile
```

```
model = Word2Vec(words, size=150, window=10, min_count=1, workers=10)
```

```
for i in words: 
  for j in words:
    if j>i:
      print(i[0],j[0],' : ',model.wv.similarity(i[0],j[0]))
    else:
      continue
```
